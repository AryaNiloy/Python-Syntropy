```python
#ODDEVEN
n = input("Enter a number: ")
if (n % 2 == 0):
    print("Even")
else:
    print("Odd")
# Greatest among three
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number:"))
num3 = int(input("Enter the third number: "))
if num1 >= num2 and num1 >= num3:
    print("The largest number is", num1)
elif num2 >= num1 and num2 >= num3:
    print("The largest number is", num2)
else:
    print("The largest number is", num3)
#Leapyear
year = int(input("Enter a year "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
#Gradecheck
marks = int(input("Enter the student's marks "))
if marks >= 90:
    print("Golden A+")
elif marks>=80 and not(marks>=90):
    print("A+")
elif marks>=70 and not(marks>=80):
    print("A-")
elif marks>=60 and not(marks>=70):
    print("B")
elif marks>=50 and not(marks>=60):
    print("B-")
elif marks>=40 and not(marks>=50):
    print("C")
elif marks>=33 and not(marks>=40):
    print("D")
else:
    print("F")
#Basic calculator
n1 = float(input("Enter the first number "))
operator = input("Enter an operator (+, -, *, /: ")
n2 = float(input("Enter the second number "))
if operator == '+':
    result = n1 + n2
elif operator == '-':
    result = n1 - n2
elif operator == '*':
    result = n1 * n2
elif operator == '/':
    result = n1 / n2
else:
    print("Invalid operator")
print("Result:", result)
#Vowelconstant        
vowels_small = 'aeiou'
vowels_big = 'AEIOU'
char = input("Enter a character ")
if char in vowels_small or char in vowels_big:
    print("Vowel")
else:
    print("Consonant")
# Thermometer
c = input("Enter 'C' for Celsius to Fahrenheit or 'F' for Fahrenheit to Celsius: ")
if c == 'C':
    celsius = float(input("Temperature in Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print("Fahrenheit is", fahrenheit)
elif c == 'F':
    fahrenheit = float(input("Enter temperature in Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"Celsius is ",celsius)
else:
    print(f"Invalid choice")
```
